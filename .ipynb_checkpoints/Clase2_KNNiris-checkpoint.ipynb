# KNN Clasificación de flores de Iris

Clasificación de flores de Iris a partir de dataset disponible en Sklearn. Se utilizará KNN para realizar clasificación categórica.

Para empezar, se realizan los imports necesarios.
1. numpy y pandas para manipulación de datos.
2. matplotlib.pyplot para visualización de los datos.
3. load_iris para cargar los datos de iris.
4. sklearn neighbors
5. sklearn.metrics, para calcular las metricas necesarias.
6. seaborn para visualizacipon de datos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn import neighbors, metrics
import seaborn as sns
```

## Entendimiento de los datos

Cargamos los datos para ver como están organizados

```python
iris = load_iris()
type(iris)
```

Los datos vienen en la clase Bunch de scikit learn. Ahora vamos a ver las variables que vienen en estos datos.

```python
print(' variable independientes: ', iris.feature_names, iris.data.shape)
print(' variables dependientes:  ', iris.target_names, iris.target.shape)
```

Lo siguiente es poner los datos en una sola estructura para manipularlos más facilmente.

```python
x = iris.data
y = iris.target
# Convertimos el arreglo en una matriz len(y) *1
y = np.expand_dims(y,axis=1)
# Concatenamos los datos
data = np.concatenate((x,y), axis=1)
data[0:5]
```

Convertimos los datos a una DataFrame para manipularlos más facilmente.

```python
# retornamos y a su forma normal para su posterior utilización
y = np.squeeze(y)
```

```python
labels = iris.feature_names.copy()
labels.append('species')
d = pd.DataFrame(data, columns=labels)
d['species'] = d['species'].astype(int).astype('str')
d.head()    
```

## Visualización de los datos

Primero vamos a visualizar las primeras dos columnas de los datos y vamos a hacer un plot de estas columans y las clasificaciones de los datos


```python
dim_x = 0
dim_y = 1
i=0
for name, group in d.groupby('species'):
    plt.scatter(x[group.index, dim_x], x[group.index, dim_y], label=iris.target_names[i])
    i+=1
plt.legend()
```

Luego, vamos a visualizar usando seaborn todas las combinaciones de estas columnas

```python
sns.pairplot(d, hue='species')
```

## Clasificación de los datos


Vamos a poner un k = 10, para que el algoritmo realiza las predicciones con respecto a los 10 elementos mas cercanos. Las variables que utilizaremos serám solo (x) como entrada y (y) como salida. El DataFrame y el procesamiento anterior se realizó con el proposito de entender los datos.

```python
K = 10
# Instanciamos el modelo
knn =  neighbors.KNeighborsClassifier(n_neighbors=K)
# Acomodamos el modelo a los datos
knn.fit(x,y)
# Realizamos las predicciones
y_pred = knn.predict(x)
print('Clases reales:    ', y)
print('Clases predichas: ', y_pred)
```
